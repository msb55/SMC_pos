```python
def flatten(array):
    return list(itertools.chain(*list(array)))
```

```python
def results_to_df(results):
    mapping = {
        'Not Pruned': '_not_pruned',
        'Pruned': '_standard',
        'Pruned Easy': '_easy',
        'Pruned Hard': '_hard',
        'Q Statistic': 'q',
        'Accuracy': 'acc',
        'G Mean': 'g_mean',
        'F-Measure': 'f1',
        'AUC': 'auc',
        'Pool size': 'size'
    }

    configurations = ['Not Pruned', 'Pruned', 'Pruned Easy', 'Pruned Hard']
    metrics = ['Q Statistic', 'Accuracy', 'G Mean', 'F-Measure', 'AUC', 'Pool size']

    dataframe = {}
    for config in configurations:
        dataframe[config] = []
        for metric in metrics:
            dataframe[config].append(results[(mapping[metric]+mapping[config])])

    return pd.DataFrame.from_dict(dataframe, orient='index', columns=metrics)

def results_to_df_mean_std(results_mean, results_std):
    mapping = {
        'Not Pruned': '_not_pruned',
        'Pruned': '_standard',
        'Pruned Easy': '_easy',
        'Pruned Hard': '_hard',
        'Q Statistic': 'q',
        'Accuracy': 'acc',
        'G Mean': 'g_mean',
        'F-Measure': 'f1',
        'AUC': 'auc',
        'Pool size': 'size'
    }

    configurations = ['Not Pruned', 'Pruned', 'Pruned Easy', 'Pruned Hard']
    metrics = ['Q Statistic', 'Accuracy', 'G Mean', 'F-Measure', 'AUC', 'Pool size']

    dataframe = {}
    for config in configurations:
        dataframe[config] = []
        for metric in metrics:
            dataframe[config].append(str(round(results_mean[(mapping[metric]+mapping[config])], 2)) + ' (' + str(round(results_std[(mapping[metric]+mapping[config])], 2)) + ')')

    return pd.DataFrame.from_dict(dataframe, orient='index', columns=metrics)

def results_to_df_mean_std2(results_mean, results_std):
    mapping = {
        'Not Pruned': '_not_pruned',
        'Pruned': '_standard',
        'Pruned Easy': '_easy',
        'Pruned Hard': '_hard',
        'Q Statistic (mean)': 'q',
        'Accuracy (mean)': 'acc',
        'G Mean (mean)': 'g_mean',
        'F-Measure (mean)': 'f1',
        'AUC (mean)': 'auc',
        'Pool size (mean)': 'size'
    }

    configurations = ['Not Pruned', 'Pruned', 'Pruned Easy', 'Pruned Hard']
    metrics = ['Q Statistic (mean)', 'Accuracy (mean)', 'G Mean (mean)', 'F-Measure (mean)', 'AUC (mean)', 'Pool size (mean)']
    metrics_std = ['Q Statistic (std)', 'Accuracy (std)', 'G Mean (std)', 'F-Measure (std)', 'AUC (std)', 'Pool size (std)']

    dataframe = {}
    for config in configurations:
        dataframe[config] = []
        for metric in metrics:
            dataframe[config].append(results_mean[(mapping[metric]+mapping[config])])
            dataframe[config].append(results_std[(mapping[metric]+mapping[config])])
            
    columns = flatten(list(zip(metrics, metrics_std)))

    return pd.DataFrame.from_dict(dataframe, orient='index', columns=columns)
```

```python
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    s_ = s.apply(lambda x: float(x[0:x.index(' (')]))
    is_max = s_ == s_.max()
    return ['font-weight: bold' if v else '' for v in is_max]

def highlight_min(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    s_ = s.apply(lambda x: float(x[0:x.index(' (')]))
    is_min = s_ == s_.abs().min()
    return ['font-weight: bold' if v else 'font-weight: normal' for v in is_min]
```

```python
def plot_compairson(first, second, metric, title, columns):
    result = pd.concat([first[[metric]], second[[metric]]], axis=1)
    result.columns = columns
    ax = result.plot.line(xlabel='Configuration', ylabel=metric, title=title, figsize=(7,5))

def plot_compairson_std(first, second, metric_mean, metric_std, title, columns, ylabel):    
    df_mean = pd.concat([first[[metric_mean]], second[[metric_mean]]], axis=1)
    df_std = pd.concat([first[[metric_std]], second[[metric_std]]], axis=1)
    df_mean.columns = columns
    df_std.columns = columns
    df_mean.plot.bar(xlabel='Configuration', ylabel=ylabel, yerr=df_std, title=title, figsize=(7,5), rot=0)
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True, ncol=len(columns))
    plt.show()
```
